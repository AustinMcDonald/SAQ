```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
dat = np.load("Ar.npy")
#Vals = [XePer, ArPer, Temp, Pres, Efield, Zdrift, Zerr,
#                Tdiff, Terr, Ldiff, Lerr, LdiffTPC, LerrTPC, TdiffTPC,
#                TerrTPC, Mele, Merr]
def gauss(x, H, A, x0, sigma):
    return H + A * np.exp(-(x - x0) ** 2 / (2 * sigma ** 2))
```

```python
Efield   = dat[:,4]
sor = np.argsort(Efield)
Efield = Efield[sor]
Vel = dat[:,5][sor]
Dt = dat[:,7][sor]
Dl = dat[:,9][sor]
```

```python
plt.plot(Efield,Vel,linewidth=3)

plt.ylim(0.9,3.4)
plt.xlim(1,299)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.grid()
plt.xlabel(r'Reduced Drift Field [V cm$^{-1}$ bar$^{-1}$]',fontsize=18)
plt.ylabel(r'Drift Velocity [mm $\mu$s$^{-1}$]',fontsize=18)
plt.tight_layout()

plt.show()
```

```python

plt.plot(Efield,Dt,linewidth=3, linestyle='--',label=r'D$_T$P')
plt.plot(Efield,Dl,linewidth=3, linestyle='--',label=r'D$_L$P')

plt.legend(loc="upper right", fontsize=18)
plt.xlim(1,299)
#plt.ylim(501,4900)
plt.grid()
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)


plt.xlabel( r'Reduced Drift Field [V cm$^{-1}$ bar$^{-1}$]',fontsize=18)
plt.ylabel(r" D P [bar cm$^2$ s$^{-1}$ ]",fontsize=18)
plt.tight_layout()
plt.show()
```

```python
plt.plot(Efield, (1.0 / (Vel/10*1e6))*1e6, label='1 cm')
plt.plot(Efield, (5.0 / (Vel/10*1e6))*1e6, label='5 cm')
plt.plot(Efield, (10.0 / (Vel/10*1e6))*1e6, label='10 cm')
plt.legend(loc="upper right", fontsize=18)

plt.xlim(0,500)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.ylabel(r"Drift Time [$\mu$s]", fontsize=18)
plt.xlabel(r"Field [V/cm]", fontsize=18)
# plt.semilogx()
# plt.axvline(0.5)
plt.show()
```
